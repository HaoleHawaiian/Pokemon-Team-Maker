```python
# imports
import pokebase as pb
import pandas as pd
```

```python
name = "swampert"
pokemon = pb.pokemon_species(name)
unique_entries = set()

for i in pokemon.flavor_text_entries:
    if i.language.name == 'en':
        flavor_text = i.flavor_text.strip().replace('\n', ' ').lower()
        unique_entries.add(flavor_text)

data = pd.DataFrame([{'Pokemon': pokemon.name.capitalize(), 'Description': text} for text in unique_entries])
print(data)
```

```python
pokemon_species = pb.pokemon_species

# Display the first Pokémon species (you can change the index to get a different species)
first_pokemon = pokemon_species(1)  # Replace 1 with the desired Pokémon ID

# Convert to a dictionary
pokemon_dict = first_pokemon.__dict__

# Display the first 5 key-value pairs
first_five_pairs = {k: pokemon_dict[k] for k in list(pokemon_dict.keys())[:5]}
print(first_five_pairs)
```
